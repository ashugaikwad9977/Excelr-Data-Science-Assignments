```python
# we used one hot encoding method for categorical data
data1 = pd.get_dummies(data)
```

```python
data1
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#we take support=0.2 to check that how many associations we got
```

```python
freq = apriori(data1,min_support=0.2,use_colnames=True)
freq
```

```python
#Now we take support=0.5 to check that how many associations we got
freq1 = apriori(data1,min_support=0.5,use_colnames=True)
freq1
```

```python
#After analyzing both supports 0.2 and 0.5 we use support 0.5
```

```python
asso = association_rules(freq1,metric='lift',min_threshold=1)
asso
```

```python
asso[asso['lift']>=1.5]
```

```python
asso.sort_values('lift',ascending=False)[0:20]
```

```python
#Now we take confidence as metrics to check the association
asso2 = association_rules(freq,metric='confidence',min_threshold=1)
asso2
```

```python
asso.sort_values('confidence',ascending=False)[0:20]
```
