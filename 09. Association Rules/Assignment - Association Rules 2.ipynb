```python
from mlxtend.frequent_patterns import apriori,association_rules
```

# if we use support=0.5 we cannot get any association

```python
freq1 = apriori(data,min_support=0.5,use_colnames=True)
freq1
```

```python
freq2 = apriori(data,min_support=0.3,use_colnames=True)
freq2
```

```python
freq3=apriori(data,min_support=0.2,use_colnames=True)
freq3
```

# if we use support=0.1 then we get association better than 0.2 so we take 0.1 as support

```python
freq4=apriori(data,min_support=0.1,use_colnames=True)
freq4
```

```python
asso = association_rules(freq4,metric='lift',min_threshold=1)
asso
```

```python
asso.sort_values('lift',ascending=False)[0:20]
```

```python
asso1 =association_rules(freq4,metric='confidence',min_threshold=0.8)
asso1
```

```python
asso.sort_values('confidence',ascending=False)[0:20]
```
