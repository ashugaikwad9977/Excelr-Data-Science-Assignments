```python
# now creating pivot table
```

```python
data_pivot=data.pivot_table(index='User.ID',
                      columns='Book.Title',
                      values='Book.Rating')
data_pivot
```

```python
data_pivot.index=data['User.ID'].unique()
data_pivot
```

```python
# as you can see there are so many nan values so we are going to fill those null values using fillna 
```

```python
data_pivot.fillna(0,inplace=True)
data_pivot
```

```python
data_pivot.shape
```

```python
# we replaced with 0 and now applying pairwise distance 
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
p_d = pairwise_distances(data_pivot.values,metric='cosine')
p_d
```

```python
rec = pd.DataFrame(p_d)
```

```python
rec.nunique()
```

```python
rec.shape
```

```python
rec.index = data_pivot.index.unique()
rec.columns = data_pivot.index.unique()
```
